```python
from PIL import Image
import matplotlib.pyplot as plt
from wordcloud import WordCloud, ImageColorGenerator
import jieba
import numpy as np
from collections import Counter
```

```python
path=r'/Users/johnson/Desktop/womany/wordcloud/'
text_file = open(path+'gossiping_300.txt', "r",encoding="utf-8").read()
```

```python
#jieba.load_userdict(r'/Users/johnson/Desktop/womany/wordcloud/stopwords.txt')
```

```python
with open(path+'gossiping_300.txt', 'r', encoding='utf8') as f:  
    stops = f.read().split('\n') 
terms = [t for t in jieba.cut(text_file, cut_all=True) if t not in stops]
sorted(Counter(terms).items(), key=lambda x:x[1], reverse=True)
```

```python
font = 'PingFang.ttc'
```

```python
mask = np.array(Image.open(r"/Users/johnson/Desktop/womany/wordcloud/car.png"))
```

```python
my_wordcloud = WordCloud(background_color="white",mask=mask,font_path=font,collocations=False, width=512, height=512, margin=2)  
my_wordcloud.generate_from_frequencies(frequencies=Counter(terms))
```

```python
plt.figure( figsize=(30,150), facecolor='k')
plt.imshow(my_wordcloud,interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
#顯示用
plt.show()
```
